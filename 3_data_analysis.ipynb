### 1.1 ¿En que momento del mes suceden las ventas?

```python
# Crear subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

# Definir una función para crear los histogramas de ventas por mes
def plot_month_histogram(df, year, ax):
    ax.hist(df['month_sale'], bins=range(1, 14), alpha=0.75, color='skyblue', edgecolor='black', align='left', rwidth=0.8)
    ax.set_title(f'Ventas por mes ({year})', fontsize=14)
    ax.set_xticks(range(1, 13))
    ax.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
    ax.set_xlabel('Mes')
    ax.set_ylabel('Número de ventas')
    ax.grid(True, linestyle='--', alpha=0.7)

# Plot para 2023
plot_month_histogram(df_2023_fs, 2023, axes[0])

# Plot para 2024
plot_month_histogram(df_2024_fs, 2024, axes[1])

plt.tight_layout()
plt.show()
```

```python
# Crear subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

# Definir una función para crear los histogramas de ventas por semana
def plot_week_histogram(df, year, ax):
    ax.hist(df['week_sale'], bins=range(1, 54), alpha=0.75, color='skyblue', edgecolor='black', align='left', rwidth=0.8)
    ax.set_title(f'Ventas por semana ({year})', fontsize=14)
    ax.set_xticks(range(1, 54, 2))
    ax.set_xlabel('Número de semana')
    ax.set_ylabel('Número de ventas')
    ax.grid(True, linestyle='--', alpha=0.7)

    # Añadir líneas de división cada 4 semanas para suponer un mes
    for i in range(4, 53, 4):
        ax.axvline(i, color='darkblue', linestyle='--', linewidth=1.5)

# Plot para 2023
plot_week_histogram(df_2023_fs, 2023, axes[0])

# Plot para 2024
plot_week_histogram(df_2024_fs, 2024, axes[1])

plt.tight_layout()
plt.show()
```

```python
# Crear subplots
fig, axes = plt.subplots(4, 3, figsize=(15, 20))

# Definir una función para crear los histogramas mensuales
def plot_histograms(df, year, axes):
    for month in range(1, 13):
        ax = axes[(month-1) // 3, (month-1) % 3]
        month_data = df[df['month_sale'] == month]
        ax.hist(month_data['day_sale'], bins=range(1, 32), alpha=0.75, color='skyblue', edgecolor='black')
        ax.set_title(f'{year} - Mes {month}')
        ax.set_xlim(1, 31)
        ax.set_xlabel('Día del mes')
        ax.set_ylabel('Número de ventas')
    fig.suptitle(f'Histograma de ventas diarias por mes ({year})', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajustar la disposición para el título principal

# Plot para 2023
plot_histograms(df_2023_fs, 2023, axes)

# Plot para 2024 en una figura nueva
fig, axes = plt.subplots(4, 3, figsize=(15, 20))
plot_histograms(df_2024_fs, 2024, axes)

plt.tight_layout()
plt.show()
```
