```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Columnas de tipos de mes
cols = ["null_month", "one_month", "two_month", "three_month", "four_month",
        "five_month", "six_month", "seven_month", "eigth_month"]

# Crear los gráficos por mes de venta
for mes in range(1, 13):
    # Filtrar las filas correspondientes a cada mes de venta
    df_mes = df[df["month_sale"] == mes]
    
    # Sumar la cantidad de ocurrencias en cada columna para ese mes
    distribucion_mes = df_mes[cols].sum()
    
    # Calcular el total de ventas en ese mes
    total_ventas_mes = distribucion_mes.sum()
    
    # Calcular los porcentajes
    distribucion_percent = (distribucion_mes / total_ventas_mes) * 100
    
    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=distribucion_percent.index, y=distribucion_percent.values, palette="viridis")
    plt.xlabel("Tipo de Mes")
    plt.ylabel("Porcentaje de Ventas (%)")
    plt.title(f"Distribución de los Tipos de Meses - Mes {mes}")
    plt.xticks(rotation=45)  # Rotar etiquetas para mejor lectura
    
    # Agregar etiquetas dentro de las barras
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width() / 2, p.get_height() + 1, 
                f'{p.get_height():.1f}%', ha='center', va='bottom', fontsize=10)
    
    plt.tight_layout()  # Ajustar el espacio
    plt.show()
```

```python
import pandas as pd

# Filtrar los valores únicos en 'program' (FS, DS, CS)
programs = ['FS', 'DS', 'CS']

# Columnas de tipos de mes
cols = ["null_month", "one_month", "two_month", "three_month", "four_month",
        "five_month", "six_month", "seven_month", "eigth_month"]

# Crear un diccionario para almacenar los resultados por programa
resultados_programas = {}

# Calcular el porcentaje por cada programa y mes de venta
for program in programs:
    resultados = []
    
    for mes in range(1, 13):
        # Filtrar las filas correspondientes a cada mes de venta y programa
        df_mes_program = df[(df["month_sale"] == mes) & (df["Program"] == program)]
        
        # Sumar la cantidad de ocurrencias en cada columna para ese mes y programa
        distribucion_mes_program = df_mes_program[cols].sum()
        
        # Calcular el total de ventas en ese mes y programa
        total_ventas_mes_program = distribucion_mes_program.sum()
        
        # Calcular los porcentajes
        distribucion_percent = (distribucion_mes_program / total_ventas_mes_program) * 100
        
        # Agregar los resultados a la lista (mes, tipo de mes, porcentaje)
        for tipo_mes, porcentaje in distribucion_percent.items():
            resultados.append([mes, tipo_mes, porcentaje])
    
    # Convertir los resultados a un DataFrame para este programa
    tabla_resultados = pd.DataFrame(resultados, columns=["Mes Venta", "Tipo de Mes", "Porcentaje"])
    
    # Pivotar la tabla para que cada tipo de mes sea una columna y cada fila sea un mes de venta
    tabla_final = tabla_resultados.pivot_table(index="Mes Venta", columns="Tipo de Mes", values="Porcentaje", aggfunc="sum")
    
    # Almacenar la tabla final por programa
    resultados_programas[program] = tabla_final

# Mostrar las tablas para cada programa
for program, tabla in resultados_programas.items():
    print(f"Tabla para el programa {program}:")
    print(tabla)
    print("\n" + "-"*50 + "\n")

```
